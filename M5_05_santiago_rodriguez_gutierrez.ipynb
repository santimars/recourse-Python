**1) Trabajemos con NumPy**

* Importa numpy y usa un alias para poder utilizar sus recursos
* Crea un array de elementos (del 10 al 100)
* Invertir el array
* Crear un array 3x3 con valores de 0 a 8
* Crear un array identidad de 6x6
* Crear un array con valores al azar con forma 3x3x3
* Crear dos arrays con valores al azar A y B, verificar si son iguales (hay varias opciones, investigar)

```python
import numpy as np

# Crear un array de elementos (del 10 al 100)
array_10_to_100 = np.arange(10, 101)
print("Array del 10 al 100:", array_10_to_100)

# Invertir el array
array_invertido = np.flip(array_10_to_100)
print("Array invertido:", array_invertido)

# Crear un array 3x3 con valores de 0 a 8
array_3x3 = np.arange(9).reshape(3, 3)
print("Array 3x3 con valores de 0 a 8:\n", array_3x3)

# Crear un array identidad de 6x6
array_identidad = np.eye(6)
print("Array identidad de 6x6:\n", array_identidad)

# Crear un array con valores al azar con forma 3x3x3
array_aleatorio = np.random.random((3, 3, 3))
print("Array al azar con forma 3x3x3:\n", array_aleatorio)

# Crear dos arrays con valores al azar A y B
A = np.random.random((3, 3))
B = np.random.random((3, 3))

# Verificar si son iguales
son_iguales = np.array_equal(A, B)
print("¿Los arrays A y B son iguales?:", son_iguales)
```

* Mostrar el resultado del array de euros pero redondeado a 2 decimales

```python
import numpy as np

# Crear un array de 100 elementos aleatorios entre 1 y 1000
dolares = np.random.randint(1, 1001, 100)
print("Array de precios en dólares:\n", dolares)

# Crear un array llamado euros
euros = dolares * 0.874413
print("Array de precios en euros:\n", euros)

# Redondear el array de euros a 2 decimales
euros_redondeados = np.round(euros, 2)
print("Array de precios en euros (redondeado a 2 decimales):\n", euros_redondeados)
```
